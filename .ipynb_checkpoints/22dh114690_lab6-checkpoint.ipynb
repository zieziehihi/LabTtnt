```python
import numpy as np

WHO_covid_data=pd.read_csv('WHO-COVID-19-global.csv')
headers=list(WHO_covid_data)
print(headers)
```

```python
every_nth_day = 1
ct_df = WHO_covid_data[WHO_covid_data[' Country'] == 'Brazil']
info=['cases']
for vr in info:
    print(vr)
    for1_df= pd.DataFrame(ct_df)
    print(for1_df.head(10))
    for1_df.plot()
    plt.show()
```

```python
every_nth_day = 1
ct_df = WHO_covid_data[WHO_covid_data[' Country'] == 'Brazil']
info=['cases']
for vr in info:
    fig,axl=plt.subplots(figsize=(20,10),dpi=50)
    color='tab:blue'
    axl.set_xlabel('days',fontsize='large',fontweight='bold')
    axl.bar(ct_df['Date_reported'][::every_nth_day],ct_df[' New_{}'.format(vr)][::every_nth_day],color=color)
    axl.tick_params(axis='y',labelcolor=color,labelsize='large')
    ax2 = axl.twinx()
    color='tab:red'
    ax2.plot(ct_df['Date_reported'][::every_nth_day],ct_df[' Cumulative_{}'.format(vr)][::every_nth_day],color=color)
    for idx ,t in enumerate(axl.get_xticklabels()):
        if(idx%5)!=0:
            t.set_visible(False)
    plt.ticklabel_format(axis='y')
    plt.xticks(rotation=90)
    dates=ct_df['Date_reported'].unique()
    start_date=dates[0]
    end_date=dates[len(dates)-1]
    plt.title('Brazil')
    fig.tight_layout()
    axl.legend(loc='upper left',fontsize=20)
    plt.setp(axl.xaxis.get_majorticklabels(),rotation=45)
    plt.show()
```
